# Matplotlib Library

Matplotlib is a python library which can be defined as multiplatform data visualization library built on numpy. We can create bar plots, scatter plots, histogram and lot more with matplotlib. It can be used in python shell, web applications and other geographical user interface toolkit. It was deleloped by John D. Hunter in 2002. It has an active community and it distributed under BSD-style license. it's first verson was release in 2003. It's latest version is 3.2.2. released on 17 June 2020.

It is the most prominent tool among Python visualization packages. Matplotlib is highly efficient in performing wide range of tasks. It can produce publication quality figures in a variety of formats. It can export visualizations to all of the common formats like PDF, SVG, JPG, PNG, BMP and GIF. It can create popular visualization types – line plot, scatter plot, histogram, bar chart, error charts, pie chart, box plot, and many more types of plot. Matplotlib also supports 3D plotting. Many Python libraries are built on top of Matplotlib. For example, pandas and Seaborn are built on Matplotlib. They allow to access Matplotlib’s methods with less code.

Matplotlib was originally started to visualize Electrocorticography (ECoG) data of epilepsy patients during post-doctoral research in Neurobiology. The open-source tool Matplotlib emerged as the most widely used plotting library for the Python programming language. It was used for data visualization during landing of the Phoenix spacecraft in 2008.

# Matplotlib Scatter

With Pyplot, we can use the scatter() function to draw a scatter plot.

The scatter() function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

We can set your own color for each scatter plot with the color or the c argument:

```python
plt.scatter(x,y, color = 'hotpink')
```

We can even set a specific color for each dot by using an array of colors as value for the c argument:

We cannot use the color argument for this, only the c argument.

```python
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors)

plt.show()
```

<h1>ColorMap</h1>

The Matplotlib module has a number of available colormaps.

A colormap is like a list of colors, where each color has a value that ranges from 0 to 100.

This colormap is called 'viridis'.

We can specify the colormap with the keyword argument cmap with the value of the colormap, in this case 'viridis' which is one of the built-in colormaps available in Matplotlib.

```python
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')
```

We can include the colormap in the drawing by including the plt.colorbar() statement:

```python
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar()
```

We can change the size of the dots with the s argument.

Just like colors, make sure the array for sizes has the same length as the arrays for the x- and y-axis:

```python
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)

plt.show()
```

<h1>Alpha</h1>

We can adjust the transparency of the dots with the alpha argument.

Just like colors, make sure the array for sizes has the same length as the arrays for the x- and y-axis:

```python
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5)
```

We can combine a colormap with different sizes of the dots. This is best visualized if the dots are transparent:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))

colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

# Matplotlib Bars

We use the bar() function to draw bar graphs:

The bar() function takes arguments that describes the layout of the bars.

The categories and their values represented by the first and second argument as arrays.


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["Python", ".NET", "Java", "C++"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

We can display bar horizontally instead of vertically, use the barh() function:

```python
plt.barh(x, y)
```

The bar() and barh() take the keyword argument color to set the color of the bars:

```python
plt.bar(x, y, color = "red")
```

The bar() takes the keyword argument width to set the width of the bars:

The default width value is 0.8

Note: For horizontal bars, use height instead of width.

```python
plt.bar(x, y, width = 0.1)
```

The barh() takes the keyword argument height to set the height of the bars:

The default height value is 0.8

```python
plt.barh(x, y, height = 0.1)
```

# Matplotlib Pie Charts

We can use the pie() function to draw pie charts:

By default the plotting of the first wedge starts from the x-axis and moves counterclockwise:

We can change the start angle by specifying a startangle parameter.

The startangle parameter is defined with an angle in degrees, default angle is 0:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([30, 35, 20, 15])

plt.pie(y)
plt.show() 
```

Add labels to the pie chart with the label parameter.

The label parameter must be an array with one label for each wedge:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Python", ".NET", "Java", "C++"]

plt.pie(y, labels = mylabels)
plt.show() 
```

```python
plt.pie(y, labels = mylabels, startangle = 90)
```

<h1>Explode</h1>

If we want one of the wedges to stand out The explode parameter allows to do that.

The explode parameter, if specified, and not None, must be an array with one value for each wedge.

Each value represents how far from the center each wedge is displayed:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Python", ".NET", "Java", "C++"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

Add a shadow to the pie chart by setting the shadows parameter to True:


```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Python", ".NET", "Java", "C++"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

<h1>Colors</h1>

We can set the color of each wedge with the colors parameter.

The colors parameter, if specified, must be an array with one value for each wedge:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Python", ".NET", "Java", "C++"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

<h1>Legend</h1>

To add a list of explanation for each wedge, use the legend() function:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Python", ".NET", "Java", "C++"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show() 
```


To add a header to the legend, add the title parameter to the legend function.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Python", ".NET", "Java", "C++"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show() 
```
